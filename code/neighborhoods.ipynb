# LA city neighborhood data

This notebook is used for examining COVID-19 cases and deaths in the Los Angeles region.

Source:
http://geohub.lacity.org/datasets/covid19-by-neighborhood-1?geometry=-120.114%2C33.645%2C-116.711%2C34.442

#### Imports

```python
import geopandas
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import contextily as ctx
```

#### Read in la city neighborhoods shapefile

```python
gdf = geopandas.read_file('../datasets/gis_data/COVID19_by_Neighborhood.shp')
gdf = gdf.to_crs(epsg=3857)
gdf.head()
```

#### Plot deaths

```python
fig, ax = plt.subplots(1, 1,figsize=(10,12))
plt.title('LA City Deaths',fontsize=24)
plt.xlabel('Longitude',fontsize=18)
plt.ylabel('Latitude',fontsize=18)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
gdf.plot(column='deaths', ax=ax, legend=True, cax=cax, cmap='Blues',edgecolor='black',alpha=0.8)
ctx.add_basemap(ax);
plt.savefig('../images/la_city_deaths.png',dpi=235,bbox_inches='tight');
```

#### Plot cases

```python
fig, ax = plt.subplots(1, 1,figsize=(10,12))
plt.title('LA City Cases',fontsize=24)
plt.xlabel('Longitude',fontsize=18)
plt.ylabel('Latitude',fontsize=18)
divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.1)
gdf.plot(column='cases', ax=ax, legend=True, cax=cax, cmap='Oranges',edgecolor='black',alpha=0.7)
ctx.add_basemap(ax)
#plt.savefig('../images/la_city_cases.png',dpi=235,bbox_inches='tight');
```
